```python


RF_features=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    RF_features[drug]=dict([(x.split()[0],float(x.split()[1])) for x in FileUtility.load_list('../../amr_results/5_3_2017/classifications_standard/'+drug+'gpa_trimmed_gpa_roaryRF_features')[1::]])
    print (drug, len(RF_features[drug]))
```

```python
Chi2_features=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    Chi2_features[drug]=dict([(x.split()[0],float(x.split()[1])) for x in FileUtility.load_list('../../amr_results/res_before_5_3_2017/results/feature_selection/chi2/'+drug+'gpa_trimmed_gpa_roary')[1::] if float(x.split()[1]) > 4])
    print (drug, len(Chi2_features[drug]))
```

```python
treewas=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    treewas[drug]=[(x.split('\t')[0].replace(' ',''),float(x.split('\t')[1])) for x in FileUtility.load_list('../../amr_results/features/treewas/gpa/'+drug+'_S.vs.R_terminal_p-vals_name.txt') if  float(x.split('\t')[1])<0.05]
    print (drug, len(treewas[drug]))
```

```python
SVM_features=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    SVM_features[drug]=[(x.split()[0],float(x.split()[1])) for x in FileUtility.load_list('../../amr_results/features/svm/'+drug+'_S-vs-R_non-zero+weights.txt')[1::]]
    print (drug, len(SVM_features[drug]))
```

```python
phylochi_features=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    phylochi_features[drug]=dict([(x.split()[0].replace('gain_','').replace('loss_',''),float(x.split()[1])) for x in FileUtility.load_list('/mounts/data/proj/asgari/dissertation/git_repos/amr_results/features/phylochi2/gpa_trimmed_gpa_roary'+drug+'.txt')[1::] if float(x.split()[1])>4])
    print (drug, len(phylochi_features[drug]))
```

```python
selected_features=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    selected_features[drug]=set(phylochi_features[drug].keys()).intersection(list(Chi2_features[drug].keys())).intersection(list(RF_features[drug].keys()))
```

```python
chi_features=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    chi_features[drug]=[x.split()[0] for x in FileUtility.load_list('../results/feature_selection/chi2/separate_drug/details/R_IS_'+drug+'_snps_all_full_trimmed.txt')[1::] if float(x.split()[1])>10]
    print (drug, len(chi_features[drug]))
```

```python
intersection_features=dict()
for drug in ['Ciprofloxacin', 'Tobramycin', 'Colistin', 'Ceftazidim', 'Meropenem']:
    intersection_features[drug]=set(set(RF_features[drug])).intersection(chi_features[drug])
    print (drug, len(intersection_features[drug]))
    FileUtility.save_list(drug+'.txt',intersection_features[drug])
```
