```python
import torch
import torch.nn as nn
import numpy as np
import torch.optim as optim
from torch.autograd import Variable

class _classifier(nn.Module):
    def __init__(self, input_size, hidden_layer, nlabel):
        super(_classifier, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(input_size, hidden_layer),
            nn.ReLU(),
            nn.Linear(hidden_layer, nlabel),
            #nn.ReLU(),            
            #nn.Linear(30, nlabel),
            nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)

def balanced_subsample(y, size=None):

    subsample = []

    if size is None:
        n_smp = y.value_counts().min()
    else:
        n_smp = int(size / len(y.value_counts().index))

    for label in y.value_counts().index:
        samples = y[y == label].index.values
        index_range = range(samples.shape[0])
        indexes = np.random.choice(index_range, size=n_smp, replace=False)
        subsample += samples[indexes].tolist()

    return subsample
```

```python
ABRAccess=ABRAccessUtility('/mounts/data/proj/asgari/github_data/data/pseudomonas/data_v2/extracted_features/',['genePA','geneexp','SNPs'])
X,Y,feature_names=ABRAccess.getXYPredictionMat_multidrug()

## dimensionality reduction
pca = PCA(n_components=1000)
X_pca = pca.fit(X.toarray())
X_new=X_pca.transform(X.toarray())

## prepare train and test
df = pd.Series( Y )
test_idx=balanced_subsample(df)
train_idx=[item for item in list(range(X.shape[0])) if item not in test_idx]
mapping={'R':1,'I':0.5,'V':0}
Y_num=[[mapping[c] for c in list(y)] for y in Y]
```

```python
# neural networ
nlabel = len(Y_num[0]) 
classifier = _classifier(X_new.shape[1], 50, nlabel)
optimizer = optim.Adam(classifier.parameters(),lr = 0.005)
criterion = nn.MultiLabelSoftMarginLoss()
```

```python
epochs = 10000
counter=0
for epoch in range(epochs):
    losses = []
    random.shuffle(train_idx)
    random.shuffle(test_idx)
    X_train=X_new[train_idx,:]
    Y_train=[Y_num[i] for i in train_idx]
    X_test=X_new[test_idx,:]
    Y_test=[Y_num[i] for i in test_idx]
    for i in range(X_train.shape[0]):
        sample=X_train[i,:]
        inputv = Variable(torch.FloatTensor(sample)).view(1, -1)
        labelsv = Variable(torch.FloatTensor(Y_train[i])).view(1, -1)
        
        output = classifier(inputv)
        loss = criterion(output, labelsv)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        losses.append(loss.data.mean())
    if counter % 100 == 0:
        print('[%d/%d] Training Loss: %.3f' % (epoch+1, epochs, np.mean(losses)))
    counter+=1
```
