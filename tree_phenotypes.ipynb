```python
import sys
sys.path.append('amrprediction/')
import codecs
import numpy as np
from data_access_utility import ABRDataAccess

```

```python
obs_tree=[l.strip().split('\t') for l in codecs.open('/Volumes/EHSAN/github/new_amr/amrprediction/data_config/GL_Ciprofloxacin.txt','r','utf-8').readlines()]
```

```python
pattern_dict=dict()

for parent,pos, pat1, pat2, phenotype in obs_tree:
    if phenotype not in pattern_dict:
        pattern_dict[phenotype]=list()
    if ' ' not in parent:
        parent=[parent]
    else:
        parent=parent.split()
    if ' ' not in pos:
        pos=[pos]
    else:
        pos=pos.split()
        
    neg=list(set(parent)-set(pos))
    pattern_dict[phenotype].append((pos,neg))
    
ABRAccess=ABRDataAccess('/Volumes/EHSAN/github/new_amr/data_v3/',['snps_all_envclin_trimmed'])
```

```python
def get_labels_feature_docs(ABRAccess, iso1_parent, iso2_iso_child):
    isolates=list(ABRAccess.isolates.values())[0]
    X=list(ABRAccess.X.values())[0]
    idx1=isolates.index(iso1_parent)
    idx2=isolates.index(iso2_iso_child)
    vec1=(X[idx1,:].toarray())
    vec2=(X[idx2,:].toarray())
    v=vec1-vec2
    loss_indx=np.where( v == 1 )[1].tolist()
    gain_indx=np.where( v == -1 )[1].tolist()
    nochange_indx=np.where( v == 0 )[1].tolist()
    loss_feature=' '.join([ABRAccess.feature_names['snps_all_envclin_trimmed'][x] for x in loss_indx])
    gain_feature=' '.join([ABRAccess.feature_names['snps_all_envclin_trimmed'][x] for x in gain_indx])
    nochange_feature=' '.join([ABRAccess.feature_names['snps_all_envclin_trimmed'][x] for x in nochange_indx])
    return loss_feature, gain_feature, nochange_feature
```

```python
i1='F1959'
i2='F1764'

a,b,c=get_labels_feature_docs(ABRAccess, i1, i2)


```

```python
import itertools

LOSS_class=[]
GAIN_class=[]
NOCH_class=[]
not_working=[]
for x_list, y_list in pattern_dict['R']:
    for x,y in list(itertools.product(x_list, y_list)):
        try:
            l,g,no_ch=get_labels_feature_docs(ABRAccess, x, y)
            LOSS_class.append(l)
            GAIN_class.append(g)
            NOCH_class.append(no_ch)
        except:
            not_working.append(y)
            not_working.append(x)
```
